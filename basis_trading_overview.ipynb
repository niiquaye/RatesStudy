Basis
- Basis = [Cash Price] - [(Futures Price) x (Conversion Factor)]
- Effects outcome of hedge
- A strengthening basis increases the cost of a short hedger 
- long hedgers benefit from a weaker basis
- keep track of historical basis to plan your hedge

Trading the Basis
- When trading the basis you must consider the following:
    - RP Specials
    - Term financing versus overnight financing 
    - Short Squeezes
    - Delivery Month
- Selling the Basis
    - Sell the bond and buy the underlying futures contract
- Buying the Basis
    - Buy the bond and sell underlying futures contract
- Hot Run Bonds (On the run bonds)
    - Very liquid
    - Bonds just issued are rich whilst bonds replaced weaken
- Selling the OTR basis is a negative carry trade

DV01
- Computed by getting the absolute value of a bond when it increases by 1% and adding it to the absolute value of a bond when it decreases by 1% then dividing by 2

What Drives the Basis
- The answer lies in the
short's rights to choose which bond to deliver and when to deliver
it. These rights make up a valuable set of strategic delivery options
that must be reflected in any bond's basis.

CTD - Cheapest to Deliver
- Calculated using either one of the following:
    - Net Basis (Lowest)
        1. Compute Basis of each security that could be delivered
        2. Compute the Carry = Coupon Income - Financing Cost
        3. Net Basis = Gross Basis - Carry
    - Implied Repo Rate (Highest)
        1. Buy Cash Bond
        2. Lend Cash bond to REPO Market
        3. Sell Bond Futures
